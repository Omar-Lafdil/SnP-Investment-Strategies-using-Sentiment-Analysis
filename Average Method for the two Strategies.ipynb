```python
def pos_or_neg(row) :
    if row["Sentiment"]< 0 :
        return (-1)
    elif row["Sentiment"]> 0:
        return 1
    else :
        return 0
```

```python
def encode(row) :
  if row["Sentiment"] < 0 :
    return -1
  elif row["Sentiment"] >0 :
    return 1
  elif row["Sentiment"] == 0 :
    return 0
```
