```python
from transformers import BertForSequenceClassification, BertTokenizer
import torch
```

```python
from transformers import pipeline
```

```python
df1 = pd.DataFrame( columns = ["articles","date", "Positive", "Negative", "Neutral"])
```

We import the FinBert model in a variable and a BertTokenizer

```python
df_list_l = []
pos_l = []
neg_l = []
neu_l = []

tokenizer = BertTokenizer.from_pretrained('ProsusAI/finbert')

model = BertForSequenceClassification.from_pretrained('ProsusAI/finbert')
```

```python

for i in range(0,19578) :
    dfi = df[10 * i:(i+1)*10]

    row = list(dfi["articles"])

    df_dates = list(dfi["date"])
    inputs = tokenizer(row, padding = True, truncation = True, return_tensors='pt') #tokenize text to be sent to model

    outputs = model(**inputs)

    predictions = torch.nn.functional.softmax(outputs.logits, dim=-1)

    model.config.id2label

    positive = predictions[:, 0].tolist()

    negative = predictions[:, 1].tolist()

    neutral = predictions[:, 2].tolist()
    
    table = {'articles':row,'date': df_dates,'Positive':positive, 'Negative':negative, 'Neutral':neutral}
    df2 = pd.DataFrame(table, columns = ["articles","date", "Positive", "Negative", "Neutral"])
  
    df1 = df1.append(df2)
    print("from :", 10 * i,"------>",(i+1)*10)
```

```python
df1 = df1.drop(columns = ["Unnamed: 0"])
```
